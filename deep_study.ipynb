```python
def forward_postproc(output, y):
    print("output : \n", output)
    print("y : \n", y)

    diff = output - y
    print("diff : \n", diff)

    square = np.square(diff)
    print("square : \n", square)

    mse = np.mean(square)
    print("mse : \n", mse)

    return mse
```
